```python
import pandas as pd
```

```python
match=pd.read_csv("./matches - matches.csv")
match.head(2)
```

```python
delivery = pd.read_csv("./deliveries.csv")
delivery.head(2)
```

```python
newDatafream=delivery.merge(match, left_on='match_id', right_on='id')
```

```python
print(delivery.shape)
print(match.shape)
```

hightest batmans runs of each season 

```python
higestScorer=newDatafream.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
higestScorer
```
