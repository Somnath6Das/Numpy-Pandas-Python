```python
def run_against(player):
    select = deliveries[deliveries['batsman']==player]
    return select.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
run_against('V Kohli')
```

```python
run_against('DA Warner')
```

```python
run_against('G Gambhir')
```

```python
run_against("MS Dhoni")
```
